```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy
```

```python
data = pd.read_csv("..//..//dataset//oakland-crime-statistics-2011-to-2016//records-for-2016.csv")
print(data)
```

#### 1. 删除空值处理

```python
#删除所有属性中有空值的行
data1 = data.dropna()
print(data1)
```

#### 2. 用最高频率值进行填充

```python
data2 = copy.deepcopy(data)#拷贝原数据集用于填充
for i in range(10):
    count = data2.iloc[:,i].value_counts()  #依次获取每一列的series
    name = pd.DataFrame(data2.iloc[:,i]).columns.values.tolist()
    sort_ans = count.sort_values(ascending = False) #对各类别的频数值进行排序 并选择最高频率的索引
    fre = sort_ans.index[0]
    fre = str(fre)
    print(name[0]+"的最高频率值为"+fre+'\n')
    
    data2[name[0]] = data2[name[0]].fillna(fre) #进行缺失值的填充
print("------------------------------------------"+'\n'+"填充完毕")
```

```python
print(data2)  #填充结果
```
