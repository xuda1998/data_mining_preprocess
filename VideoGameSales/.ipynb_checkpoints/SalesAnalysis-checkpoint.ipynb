> + 对整个销售情况进行相应的的可视化

## 一、电子游戏市场分析

```python
#导入需要用到的包，包括numpy，pandas以及sklearn等
import numpy as np
import pandas as pd
from pyecharts import options as opts
from sklearn.naive_bayes import GaussianNB
from pyecharts.charts import Bar,Pie
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("vgsales.csv")
data.head()
```

```python
data.info()
```

```python
#这里我们做一下简单的缺失值处理
data.dropna(inplace = True)
data.reset_index(drop=True,inplace = True)
data.head(10)
```

### 受欢迎的游戏、类型、发行平台、发行人

```python
sort_data = data.sort_values("Global_Sales",inplace=False,ascending = False)
show_data = sort_data.head(20) #排名前20的游戏
```

```python
bar1 = Bar(init_opts=opts.InitOpts(width='1450px', height='350px'))
bar1.add_xaxis(show_data['Name'].tolist())
bar1.add_yaxis("",show_data['Global_Sales'].tolist())
bar1.set_global_opts(title_opts=opts.TitleOpts(title="1980-2020年最受欢迎的游戏"),
                     visualmap_opts=opts.VisualMapOpts(max_=100),
                    ) 
bar1.render_notebook()
```

```python
bar1 = Bar(init_opts=opts.InitOpts(width='1450px', height='350px'))
bar1.add_xaxis(sort_data['Genre'].value_counts().index.tolist())
bar1.add_yaxis("",sort_data['Genre'].value_counts().tolist())
bar1.set_global_opts(title_opts=opts.TitleOpts(title="1980-2020年用户喜欢的游戏类型"),
                     visualmap_opts=opts.VisualMapOpts(max_=5000),
                    ) 
bar1.render_notebook()
```

```python
bar1 = Bar(init_opts=opts.InitOpts(width='1450px', height='350px'))
bar1.add_xaxis(sort_data['Platform'].value_counts().index.tolist())
bar1.add_yaxis("",sort_data['Platform'].value_counts().tolist())
bar1.set_global_opts(title_opts=opts.TitleOpts(title="1980-2020年游戏平台用户量"),
                     visualmap_opts=opts.VisualMapOpts(max_=2500),
                    ) 
bar1.render_notebook()
```

```python
pie1 = Pie(init_opts=opts.InitOpts(width='1350px', height='550px'))
pie1.add('', [list(z) for z in zip(sort_data['Publisher'].value_counts().head(10).index.tolist(), sort_data['Publisher'].value_counts().head(10).tolist())], radius=['35%', '60%'])
pie1.set_global_opts(title_opts=opts.TitleOpts(title='1980-2020全球前十发行商'), 
                     legend_opts=opts.LegendOpts(orient='vertical', pos_top='15%', pos_left='2%'))
pie1.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{d}%"))
pie1.render_notebook()
```

## 二、全球电子游戏销售额预测 

```python
region = ['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
region_sales_sum =pd.pivot_table(data,index='Year',values=region,aggfunc='sum')
plt.figure(figsize=(15,8))
sns.lineplot(data=region_sales_sum)
plt.title('地区销量发展趋势')
```
