```python
本实验选取了wine-reviews数据集作为实验样本，本文档主要是对该数据集中的标称属性进行分析并获取其各聚会的频数
```

```python
import csv
import numpy as np
import pandas as pd
```

```python
#读取数据文件
data = pd.read_csv("..\..\dataset\wine-reviews\winemag-data_first150k.csv",usecols=['country','description','designation','province','region_1','region_2','variety','winery'])
```

```python
count = data.loc[:,'country'].value_counts()  #获取对应标称属性列的series并进行计数统计 以下各段代码同理
print(count)
```

```python
count = data.loc[:,'description'].value_counts()
print(count)
```

```python
count = data.loc[:,'designation'].value_counts()
print(count)
```

```python
count = data.loc[:,'province'].value_counts()
print(count)
```

```python
count = data.loc[:,'region_1'].value_counts()
print(count)
```

```python
count = data.loc[:,'region_2'].value_counts()
print(count)
```

```python
count = data.loc[:,'variety'].value_counts()
print(count)
```

```python
count = data.loc[:,'winery'].value_counts()
print(count)
```
